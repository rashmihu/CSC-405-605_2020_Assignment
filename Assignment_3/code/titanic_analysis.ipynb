```python
# you would need xlrd - pip install xlrd
t_file = pd.ExcelFile('../data/titanic.xls')
t_df = t_file.parse("titanic", header=None)
t_df.head(20)
```

### Women and children first?

*** 1. Use the `groupby` method to calculate the proportion of passengers that survived by sex. (10 pts)*** 

```python
# Rename the Column headers with the 0th row data
t_df= t_df.rename(columns=t_df.iloc[0])
t1_df = t_df.iloc[1:]  
t1_df
```

### Filter the data based on the survival and get the count of total number of people who survived

```python
#Collect the sex column data for all the people who survived
df_survived = t1_df[t1_df['survived'] == 1]['sex'].reset_index()
#Calculate the the total count of total number of people who survived 
total_survived = t1_df[t1_df['survived'] == 1]['survived'].count();
total_survived
```

### Groupby sex and calculate the proportions

```python
#Group by sex column data for all the people who survived and calculate the proportion
df_groupedBySex= df_survived.groupby('sex').count().apply(lambda x:
                                                 100 * x / float(total_survived)).reset_index()
df_groupedBySex
```

*** 2. Calculate the same proportion, but by class and sex. (10 pts)*** 

### Groupby sex and class and  then calculate the proportions

```python
# Collect the sex and pclass column data for all the people who survived
df_survived = t1_df[t1_df['survived'] == 1][['sex','pclass']].reset_index()
# Group by sex and class column data for all the people who survived and calculate the proportion
df_groupedBySexAndPclass= df_survived.groupby(['sex','pclass']).count().apply(lambda x:
                                                 100 * x / float(total_survived)).reset_index()
df_groupedBySexAndPclass
```

*** 3. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. (20 pts)***

```python
# Filter the data based on the age categories 
conditions =[(t1_df['age'].between(0,14)),
             (t1_df['age'].between(14,20,inclusive=True)),
             (t1_df['age'].between(21,64,inclusive=True)),
             (t1_df['age']>=65)
    
] 
# Assign values to each category
values = ['chidren','adolescents','adult','senior']
#Create a new data frame with the values based on the category conditions
t1_df['ageCategory'] = np.select(conditions,values)    
t1_df
```

### Groupby agecategory,class and sex and then calculate the proportion of people who survived

```python
# Groupby agecategory,class and sex
df_survived = t1_df[t1_df['survived'] == 1][['ageCategory','pclass','sex']].reset_index()
# Calculate the proportion based on the agecategory,class and sex
df_groupedByAgeCategorySexAndPclass= df_survived.groupby(['ageCategory','pclass','sex']).count().apply(lambda x:
                                                 100 * x / float(total_survived)).reset_index()
df_groupedByAgeCategorySexAndPclass
```
